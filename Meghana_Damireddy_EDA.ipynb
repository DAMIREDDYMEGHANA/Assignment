```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the datasets
customers=pd.read_csv("C:\\Users\\meghana\\Downloads\\Customers.csv")
products=pd.read_csv("C:\\Users\\meghana\\Downloads\\Products.csv")
transactions=pd.read_csv("C:\\Users\\meghana\\Downloads\\Transactions.csv")
```

```python
#cleaning the data
print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())
```

```python
print(customers.duplicated().sum())
print(products.duplicated().sum())
print(transactions.duplicated().sum())
```

```python
customers['Region'].value_counts().plot(kind='bar')
plt.title('Customers by Region')
plt.show()
```

```python
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
customers['SignupDate'].dt.year.value_counts().sort_index().plot(kind='line')
plt.title('Signups Over Time')
plt.show()
```

```python
products['Category'].value_counts().plot(kind='bar')
plt.title('Products by Category')
plt.show()
```

```python
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
transactions.groupby(transactions['TransactionDate'].dt.date)['TotalValue'].sum().plot(kind='line')
plt.title('Total Sales Over Time')
plt.show()
```
